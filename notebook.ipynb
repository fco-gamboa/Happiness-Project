```python
import pandas as pd

df = pd.read_csv('world-happiness-report-2021.csv')
df.head()
```

```python
rename_dict = {
    'Country name': 'country',
    'Regional indicator': 'region',
    'Ladder score': 'ladder_score',
    'Standard error of ladder score': 'ladder_score_se',
    'upperwhisker': 'upper_whisker',
    'lowerwhisker': 'lower_whisker',
    'Logged GDP per capita': 'gdp_per_capita_log',
    'Social support': 'social_support',
    'Healthy life expectancy': 'life_expectancy',
    'Freedom to make life choices': 'freedom',
    'Generosity': 'generosity',
    'Perceptions of corruption': 'corruption',
    'Ladder score in Dystopia': 'ladder_score_dystopia',
    'Explained by: Log GDP per capita': 'exp_gdp',
    'Explained by: Social support': 'exp_social_support',
    'Explained by: Healthy life expectancy': 'exp_life_expectancy',
    'Explained by: Freedom to make life choices': 'exp_freedom',
    'Explained by: Generosity': 'exp_generosity',
    'Explained by: Perceptions of corruption': 'exp_corruption',
    'Dystopia + residual': 'dystopia_residual'
}

# Renommage
df = df.rename(columns=rename_dict)
# Convertir en valeurs numerique si applicable
df = df.apply(pd.to_numeric, errors='ignore')
# Efacer les espaces avant et apres
df['country'] = df['country'].str.strip()
# enlever valeur manquants
df = df.dropna()   

df.info()
```
